#**Functions:**

function is a block of code that can be reused in a program

Factorial of a number

it is the product of all the positive integers less than or equal to the given number

Factorial of 5=5*4*3*2*1=120

```python
number=int(input("Enter a number to find its factorial:"))
factorial=1;
if number==0:
  print("The factorial of 0 is 1")
else:
  for i in range(1,number+1): #range=(1,5+1)->(1,6)-->1,2,3,4,5
    factorial=factorial*i

  print('The factorial of ',number,'is',factorial)
```

```python
#how to create factorial program using functions
def factorial_value(num):
  factorial=1;
  if num==0:
   return factorial
  else:
   for i in range(1,num+1):
    factorial=factorial*i
   return factorial
```

```python
print(factorial_value(5))
```

```python
print(factorial_value(10))
```

```python
#last five rows of dataframe
credit_card_data.tail()
```

```python
credit_card_data.shape
```

```python
#distribution of the two classes
credit_card_data['Class'].value_counts()
```

This is the highly imbalenced dataset

0-->legit transactions

1-->Fraudulent transactions

```python
#separating the legit and fradulent transactions
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

UnderSampling

to handle imbalenced data we use this techinqiue

Building the sample dataset containing similar distribution of Legit and Fraudulent transactions

Number of fraudulent transactions:2

```python
legit_sample=legit.sample(n=2)
```

```python
print(legit_sample.shape)
```

Concatinate the two dataframes

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```
